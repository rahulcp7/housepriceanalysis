```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns

##display all the columns of the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('train.csv')

##print shape of dataset with rows and columns
print(dataset.shape)
```

```python
## print the top 5 rows
dataset.head()
```

```python
#Check the percentage of null values(Empty) for each feature
# Step 1 - Make the list of features with empty values

features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

#Step 2 - Print the featurename and % of null values

for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),"% missing values")
```

### Since there are missing values, we need to find their relationship with Sales

```python
for feature in features_with_na:
    data = dataset.copy()
    #Lets make a variable that indicates 1 if the observation was missing or 0 if value available
    data[feature]=np.where(data[feature].isnull(),1,0)

    #Lets calculate the mean sales price where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

#### Clearly there is a dependency between null values and SalePrice feature and hence null values has to be replaced with some meaningful values in Feature Engineering

##### From the above dataset, features like IDs are not required

```python
print("IDs of houses {}".format(len(dataset.Id)))
```

### Numerical Variables

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
print('Number of numerical variables: ',len(numerical_features))

dataset[numerical_features].head()
```

### Temporal variable
##### Features which have time related data like Datetime, Year etc

```python
# list the features which have year information
year_features=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_features
```

```python
#Lets explore content of these year variables
for feature in year_features:
    print(feature,dataset[feature].unique())
```

### Analysis of Temporal features

```python
# We will check if there was a relationship between the year house is sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs Year Sold")
```

```python
# We will compare the difference between all year features with Sale price relationships
for feature in year_features:
    if feature!='YrSold':
        data=dataset.copy()
        # We will capture the difference betweeny 
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

### Discrete Variables

```python
discrete_features=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_features+['Id']]

print('Total Discrete features:',len(discrete_features))
discrete_features
```

```python
dataset[discrete_features].head()
```

### Let us find the relationship between discrete features and Sale Price

```python
for feature in discrete_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()

```

#### There is a relationship between Discrete features and Sale price. Now let us check the relationship between continuous variables and Sale Price

```python
conti_features=[feature for feature in numerical_features if feature not in discrete_features+year_features+['Id']]
print('Total continuous features:',len(conti_features))
```

#### For continuous values we need to find out distribution of continuous values and for that we have to plot histograms

```python
for feature in conti_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

## Exploratory Data Analysis Part 2

```python
# We will be using logarithmic transformation


for feature in conti_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.title(feature)
        plt.show()
```

### Outliers

```python
for feature in conti_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
